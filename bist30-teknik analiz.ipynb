```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt


bist_30= ["AKBNK.IS", "ARCLK.IS", "ASELS.IS", "BIMAS.IS", "DOHOL.IS", 
    "EKGYO.IS", "EREGL.IS", "FROTO.IS", "GARAN.IS", "GOZDE.IS", 
    "HALKB.IS", "HEKTS.IS", "ISCTR.IS", "KCHOL.IS", "KOZAA.IS", 
    "KOZAL.IS", "MGROS.IS", "PETKM.IS", "PGSUS.IS", "SAHOL.IS", 
    "SASA.IS", "SISE.IS", "SKBNK.IS", "TAVHL.IS", "THYAO.IS", 
    "TKFEN.IS", "TOASO.IS", "TSKB.IS", "TTKOM.IS", "TUPRS.IS"]

for sirket in bist_30:
    data = yf.download(sirket, start="2024-01-01", end="2024-05-28")
    data["20_gun_ort"] = data["Close"].rolling(window=20).mean()
    data["50_gun_ort"] = data["Close"].rolling(window=50).mean()
    

    plt.figure(figsize=(8, 6))
    plt.plot(data["Close"], label="Kapan ")
    plt.plot(data["20_gun_ort"], label="20 Gnlk ort",)
    plt.plot(data["50_gun_ort"], label="50 Gnlk ort")
    plt.title(f"{sirket} Teknik Analizi")
    plt.legend()
    plt.show()
    
 
```
